# Monitorando vendas no comércio

Programa que recebe o número de vendas dos dois produtos e exibe uma mensagem indicando qual deles vendeu mais. Se as quantidades forem iguais, exibe uma mensagem dizendo que houve empate.

```python
produto_1 = int(input('Digite o número de vendas do produto 1:  '))
produto_2 = int(input('Digite o número de vendas do produto 2:  '))

if produto_1 > produto_2:
  print('O produto 1 vendeu mais')

elif produto_2 > produto_1:
  print('O produto 2 vendeu mais')

else:
  print('Houve empate')
```

# Cálculo de IMC

O programa deve receber o peso e a altura de uma pessoa e exibir o valor do IMC, além de indicar se está abaixo do peso, com peso normal ou acima do peso.

```python
peso = float(input('Digite o peso: '))
altura = float(input('Digite a altura: '))

imc = peso / altura ** 2
print('Seu IMC é:', imc)

if imc < 18.5:
  print('Abaixo do peso')

elif imc >= 18.5 and imc <= 24.9:
  print('Peso normal')

else:
  print('Acima do peso')
```

# Controle do orçamento mensal

O programa deve receber o total de despesas realizadas e informar se ele ultrapassou o limite de R$3.000,00 ou ainda está dentro do orçamento.

```python
alimentacao = float(input('Digite o total de gastos com alimentação: '))
moradia = float(input('Digite o total de gastos com moradia: '))
transporte = float(input('Digite o total de gastos com transporte: '))
outros = float(input('Digite o total de gastos com outros: '))
total = alimentacao + moradia + transporte + outros

print('O total de gastos é:', total)

if total > 3000:
  print('Você ultrapassou o orçamento')

else:
  print('Você está dentro do orçamento')
```

# Cálculo de pedágio

O valor do pedágio depende da distância percorrida:

Até 100 km: R$ 10,00

Entre 100 km e 200 km: R$ 20,00

Acima de 200 km: R$ 30,00

O programa recebe a distância percorrida e informa o valor do pedágio correspondente.

```python
distancia = float(input('Digite a distância percorrida: '))

if distancia <= 100:
  print('O valor do pedágio é: R$ 10,00')

elif distancia < 100 and distancia <= 200:
  print('O valor do pedágio é: R$ 20,00')

else:
  print('O valor do pedágio é: R$30,00')
```

# Aprovar empréstimo

Um cliente quer solicitar um empréstimo, mas a aprovação depende de duas condições:

O valor da renda mensal precisa ser maior que R$ 2.000,00.
O valor da parcela não pode ultrapassar 30% da renda.
O programa recebe como entrada a renda mensal do cliente e o valor da parcela desejada. O programa informa se o empréstimo foi aprovado ou negado com base nas condições acima.

```python
renda_mensal = float(input('Digite a renda mensal: '))
valor_parcela = float(input('Digite o valor da parcela: '))

if renda_mensal <= 2000:
  print('Empréstimo negado')

elif valor_parcela > 0.3 * renda_mensal:
  print('Empréstimo negado')

else:
  print('Empréstimo aprovado')
```
